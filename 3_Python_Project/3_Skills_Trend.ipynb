# What are the trends of in-demand skills for Data Analyst?

**Methodology**:
1. Aggregating skill counts by months
2. Re-analyzing based on the percentage of total job postings
3. Plotting the monthly skill demand

```python
#importing libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import seaborn as sns


#loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#quick cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
#filtering for only Data Analyst roles in the United States
df_DA_US = df[(df['job_title'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
```

```python
#creating a column of month number for easier filtering
df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.month
```

```python
#separating the job_skills into separate rows with the same index
df_DA_US_explode = df_DA_US.explode('job_skills')
```

```python
df_DA_US_pivot = df_DA_US_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

df_DA_US_pivot.loc['Total'] = df_DA_US_pivot.sum() #creating a new row of total for sorting

df_DA_US_pivot = df_DA_US_pivot[df_DA_US_pivot.loc['Total'].sort_values(ascending=False).index]

df_DA_US_pivot = df_DA_US_pivot.drop('Total') #dropping the 'total' row since it would skew our graphs
```

```python
#getting the total values of postings per skill
DA_totals = df_DA_US.groupby('job_posted_month_no').size()
```

```python
#dividing the skills postings per month by the total value of DA_totals
df_DA_US_percentage = df_DA_US_pivot.div(DA_totals/100, axis=0)
```

```python
#changing the month number to month name for better visuals
df_DA_US_percentage = df_DA_US_percentage.reset_index()
df_DA_US_percentage['job_posted_month'] = df_DA_US_percentage['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_US_percentage = df_DA_US_percentage.set_index('job_posted_month')
df_DA_US_percentage = df_DA_US_percentage.drop(columns='job_posted_month_no')

df_DA_US_percentage
```

```python
df_plot = df_DA_US_percentage.iloc[:, :5]

sns.lineplot(data=df_plot, dashes=False, palette='dark:salmon_r')
sns.set_theme(style='ticks')
sns.despine()

plt.title('The Trends for Top Skills of Data Analysts in the US')
plt.ylabel('Likelihood in Job Postings')
plt.xlabel('2023')
plt.legend().remove()

from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))

for i in range(5):
    plt.text(11.2, df_plot.iloc[-1, i], df_plot.columns[i])
```
