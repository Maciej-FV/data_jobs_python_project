```python
#importing libraries

import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#quick cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

### Index
---

```python
df.index.name = 'job_index'
```

```python
df
```

```python
df_poland = df[df['job_country'] == 'Poland']

df_poland
```

```python
df_poland.reset_index(inplace=True)
df_poland
```

```python
df_poland.set_index('job_index', inplace=True)
df_poland
```

```python
median_pivot = df_poland.pivot_table(
    values='salary_year_avg', 
    index='job_title_short', 
    aggfunc=['median','min','max'])

median_pivot
```

```python
median_pivot.sort_values(by=('median', 'salary_year_avg'), ascending=False)
```
