```python
#importing libraries

import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#quick cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_pl = df[df['job_country'] == 'Poland'].copy()

df_pl
```

```python
df_pl['job_posted_month'] = df_pl['job_posted_date'].dt.strftime('%B')
df_pl
```

```python
df_pl_pivot = df_pl.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
```

```python
df_pl_pivot.reset_index(inplace=True)
df_pl_pivot['month_no'] = pd.to_datetime(df_pl_pivot['job_posted_month'], format ='%B').dt.month
df_pl_pivot.sort_values('month_no', inplace=True)
df_pl_pivot.set_index('job_posted_month', inplace=True)
df_pl_pivot.drop(columns='month_no', inplace=True)
df_pl_pivot
```

```python
df_pl_pivot.plot(kind='line')
```

```python
top_3 = df_pl['job_title_short'].value_counts().head(3)
top_3 = top_3.index.tolist()
```

```python
df_pl_pivot[top_3].plot(kind='line')
plt.title('Monthly JobPostings for Top 3 Data Titles in Poland')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.legend()
plt.show()
```

```python
df_pl_software_pivot = pd.read_csv("https://lukeb.co/software_csv", index_col='job_posted_month')
df_pl_software_pivot
```

```python
df_pl_merged = df_pl_pivot.merge(df_pl_software_pivot, on='job_posted_month')
```

```python
top_5 = (
    df_pl_merged
    .sum()
    .sort_values(ascending=False)
    .head()
    .index
    .to_list())

top_5
```

```python
#the plot is quite inaccurate due to the fact that Developer Data is fictitious data set
# and ruins the representation of data jobs with the higher values (as they are prepared for the USA job market) 
df_pl_merged[top_5].plot(kind='line')
plt.title('Monthly Job Postings for Top Tech Jobs in Poland')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.ylim(0,20000)
plt.legend()
plt.show()
```
