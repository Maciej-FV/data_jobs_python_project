```python
#importing libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt


#loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#quick cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie', startangle=90,autopct='%1.1f%%') 
#startangle is self-explanatory, autopct = autopercentage - % means formatting if i understand correctly, 1 is the number, . is the period, 1 after period is number of decimal places, %% makes the % sign
plt.title('Work from Home Status')
plt.ylabel('')
plt.show()
```

```python
fig, ax = plt.subplots(1, 3)

dict_column = {
    'job_work_from_home': 'Work From Home',
    'job_no_degree_mention': 'Job Degree Req.',
    'job_health_insurance': 'Health Insurance Offered' 
}

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df[column].value_counts(), startangle=90, autopct='%1.1f%%', textprops={'fontsize':9}, labels=['False', 'True'])
    ax[i].set_title(title, fontsize=10)
    
```
